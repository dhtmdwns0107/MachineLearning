## Tokenizer

```python
# stopwords -> 한국어의 조사 은, 는, 이, 가, 을, 를, 잘, 도...
stopwords=['의', '가', '이', '는', '은', '들', '잘', '좀', '와',
           '과', '도', '을', '를', '으로', '한', '하다', '에'] 
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
okt.morphs("2000년 이후 최고의 서스팬스 영화.. 라고 했지만, 이걸 왜 봤을까? ", stem=True)
```

```python
# 불용어 제거, 형태소 토근화 작업
X_train = []
for sentence in train_data['document']:
    temp_x = []
    temp_x = okt.morphs(sentence, stem=True) # Tokeniziing
    temp_x = [word for word in temp_x if not word in stopwords] # stop word
    X_train.append(temp_x)
print(X_train[:5])
```

```python
X_test = []
for sentence in test_data['document']:
    temp_x = []
    temp_x = okt.morphs(sentence, stem=True) # Tokeniziing
    temp_x = [word for word in temp_x if not word in stopwords] # stop word
    X_test.append(temp_x)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
print(tokenizer.word_index)
# token_size or vector_size
```

```python
threshold = 3 # 빈도수 (3회 미만 무시)
total_cnt = len(tokenizer.word_index) # 단어수
rare_cnt = 0 # 빈도수가 threshold보다 작은 단어수
total_freq = 0 # 훈련 데이터 전체 단어 빈도수 총 합
rate_freq = 0 # 빈도수가 threshold보다 작은 단어들의 총 합

# 단어와 빈도수의 pair를 key, value로 처리 
for key, value in tokenizer.word_index.items():
    total_freq = total_freq + value 
    
    # 단어 빈도수가 threshold보다 작은 값 처리 
    if (value < threshold):
        rare_cnt += 1
        rate_freq += value 
```

```python
print('총 단어 수(Vocabulary):', total_cnt)
print('빈도수 낮은 수:', rare_cnt)
print('단어 집합에서 빈도수가 낮은 단어 비율:', 
      ((rare_cnt / total_cnt) * 100))
print('전체 출현 단어에서 빈도수가 낮은 단어 비율:',
      ((rate_freq / total_freq) * 100))

# 전체 단어 수 중 빈도구가 3이하인 단어 제거, 
vocab_size = total_cnt - rare_cnt + 1
print('단어 집합의 크기:', vocab_size)
```

```python
tokenizer = Tokenizer(vocab_size)
tokenizer.fit_on_texts(X_train)
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
print(X_train[:3]) # [[150, 12, 434, 5, 454, ...]]
```

```python
Y_train = np.array(train_data['label']) 
Y_test = np.array(test_data['label']) 
```

```python
# 빈 샘플 데이터 제거 
```

```python
# 패딩 처리 
# 최대 리뷰 길이, 평규 리뷰 길이 구하기 
max_review_length = max(len(x) for x in X_train)
print("최대 리뷰 길이: ", max_review_length)
# ave, mean ...
mean_review_length = sum(map(len, X_train)) / len(X_train)
print("평균 리뷰 길이: ", mean_review_length)
plt.hist([len(s) for s in X_train], bins=50)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
```

```python
max_length = 30
X_train = pad_sequences(X_train, maxlen=max_length)
X_test = pad_sequences(X_test, maxlen=max_length)
```

## LSTM을 이용한 리뷰 모델 생성 

```python
from tensorflow.keras.layers import Embedding, Dense, LSTM, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
T = X_train.shape[1] # 30
i = Input(shape=(T,))
x = Embedding(vocab_size, 100)(i) # V? D? 
x = LSTM(128)(x)
x = Dense(1, activation='sigmoid')(x)
model = Model(i, x)
model.summary()
```

```python
eStop = EarlyStopping(monitor='val_loss', mode='min', 
                     verbose=1, patience=5)
mCheck = ModelCheckpoint('best_model.h5', monitor='val_acc',
                        mode='max', verbose=1,
                        save_best_only=True)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=10, validation_split=0.2,
         batch_size=60, callbacks=[eStop, mCheck])
```

```python
print("테스트 정확도:", model.evaluate(X_test, Y_test)[1])
# 테스트 정확도: 0.8xxx 
```
